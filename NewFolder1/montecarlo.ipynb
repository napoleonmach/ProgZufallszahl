```python
import random
import math

approxPi = 0
iterations = 0
inCircle = 0

while abs(approxPi - math.pi) > 1e-5:
    x = random.random()
    y = random.random()
    iterations += 1

    if math.sqrt(x * x + y * y) < 1:
        inCircle += 1

    approxPi = (inCircle / iterations) * 4.0

    print(f"Iteration {iterations}: {approxPi}")

print(f"Ende: {approxPi}")
```
